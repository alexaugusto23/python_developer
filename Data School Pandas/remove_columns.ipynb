# Como remover colunas de DataFrame.

```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.head(2)
```

```python
ufo.shape
```

```python
ufo.drop('Colors Reported', axis=1, inplace=True)
```

```python
ufo.head(2)
```

```python
# Deletando colunas
ufo.drop(['City', 'State'], axis=1, inplace=True)
ufo
```

```python
# Deletando linhas
ufo.drop( [0,1], axis=0, inplace=True)
ufo
```

```python
# Usando del para deletar colunas
del ufo['Time']
ufo
```
