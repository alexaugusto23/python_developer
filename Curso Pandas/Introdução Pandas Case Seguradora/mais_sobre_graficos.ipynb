## Explorando funcionalidades gráficas

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt 
plt.rc('figure', figsize = (15, 8))
```

```python
dados = pd.read_csv('dados/aluguel_residencial_sem_outliers.csv', sep = ';')
dados.head()
```

```python
area = plt.figure()
```

```python
g1 = area.add_subplot(2, 2, 1)
g2 = area.add_subplot(2, 2, 2)
g3 = area.add_subplot(2, 2, 3)
g4 = area.add_subplot(2, 2, 4)
```

```python
dados_g3 = dados['Valor'].sample(100)
dados_g3.index = range(len(dados_g3)) # dados_g3.shape[0]
dados_g3
```

```python
grupo = dados.groupby('Tipo')['Valor']
grupo.groups
```

```python
label = grupo.mean().index
valores = grupo.mean().values
```

```python
g1.scatter(dados['Valor'], dados['Area'])
g1.set_title('Valor X Area')

g2.hist(dados['Valor'])
g2.set_title('Histograma')

g3.plot(dados_g3)
g3.set_title('Amostra (Valor)')

g4.bar(label, valores)
g4.set_title('Valor Médio por tipo')
```

```python
graph_imoveis = plt.figure()
```

```python
graph1 = graph_imoveis.add_subplot(1,2,1) 
graph2 = graph_imoveis.add_subplot(1,2,2) 
```

```python
grupo1 = dados.groupby('Tipo Agregado')['Valor']
grupo1.groups
label = grupo1.count().index
print(label)
valores = grupo1.count().values
print(valores)
```

```python
grupo2 = dados.groupby('Tipo')['Valor']
grupo2.groups
label_g2 = grupo2.count().index
print(label)
valores_g2 = grupo2.count().values
print(valores)
```

```python
graph1.pie(valores, labels = label, autopct='%1.1f%%')
graph1.set_title('Total de Imóveis por tipo Agregado')

graph2.pie(valores_g2, labels = label_g2, autopct='%1.1f%%', explode = (.1, .1, .1, .1, .1))
graph2.set_title('Total de Imóveis por Tipo')
```
