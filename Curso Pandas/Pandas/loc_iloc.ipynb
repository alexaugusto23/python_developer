# Selecionando multiplas linhas e colunas com LOC

```python
# Importando Pandas 
import pandas as pd
```

```python
# Lendo dataset
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head(3)
```

```python
# Loc filtra linhas colunas by labels exemplo index ou colunas
# [linhas, colunas]
# [inicio:fim , inicio:fim]
# [: , :] tudo
ufo.loc[0, :] # linha 0 e todas as colunas
ufo.loc[0] # linha 0 e todas as colunas
```

```python
# Passando as linhas como lista.
ufo.loc[[0,1,2], :] # linha 0 e todas as colunas
ufo.loc[0:2, :] # linha 0 e todas as colunas
```

```python
# Tornando a coluna de resultado explicita.
ufo.loc[: , 'City'] # linha 0 e todas as colunas
```

```python
ufo.loc[: , ['City','State']] 
```

```python
# Criando um range das colunas.
ufo.loc[0:2 , 'City':'State'] 
```

```python
ufo.head(3).drop('Time', axis=1)
```

```python
ufo[ufo['City'] == 'Oakland']
ufo[ufo.City == 'Oakland']
```

```python
ufo.loc[ufo['City'] == 'Oakland', :]
ufo.loc[ufo.City == 'Oakland', :]
```

```python
ufo.loc[ufo['City'] == 'Oakland'].State
```

```python
ufo.loc[ufo['City'] == 'Oakland', 'State']
```

# Selecionando multiplas linhas e colunas com ILOC utilizando sua posiÃ§Ã£o no dataframe.

```python
# Selecionando pela posiÃ§Ã£o
#[Linhas, Colunas]
# posiÃ§Ã£o [:, [0,3]]
ufo.iloc[:, [0,3]]
```

```python
ufo.columns
```

```python
# Fazendo um range para colunas
ufo.iloc[:, 0:4]
```

```python
# Fazendo um range para colunas
ufo.iloc[0:1, :]
```

```python
# SeleÃ§Ã£o de colunas 
ufo[['City', 'State']]
ufo.loc[ : , ['City', 'State']]
```

```python
# SeleÃ§Ã£o de linhas
ufo[0:2]
ufo.iloc[0:2 , :]
```
